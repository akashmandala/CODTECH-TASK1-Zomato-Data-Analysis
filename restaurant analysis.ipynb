```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df
```

```python
import seaborn as sns
```

```python
#Data cleaning
```

```python
def handleRate(value):
    value = str(value).split("/")
    value = value[0];
    return float(value)
df['rate']=df['rate'].apply(handleRate)
print(df.head())
    
```

```python
df.info()
```

```python
#Finding which type of restaurants have more customers
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of Restaurant")
```

```python
#conclusion : Dining type of restaurants have more customers
```

```python
#Finding how many votes each type of restaurant has received from customers
```

```python
grouped_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c='yellow',marker="o")
plt.xlabel("type of Restaurant",c="yellow",size=20)
plt.ylabel("votes",c="yellow",size=20)
```
